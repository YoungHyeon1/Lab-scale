```python
players = []
for player in test_data['participants']:
    players.append(player['championId'])
    players.append('DIAMOND')
    players.append(player['teamId'])
    players.extend(player['perks']['perkIds'][:6])
    # players.extend([, player['perks']['perkSubStyle']])
    players.insert(7, player['perks']['perkStyle'])
    players.append(player['perks']['perkSubStyle'])
    # players.insert(0, player['perks']['perkSubStyle'])
    # players.append( player['perks']['perkIds'])
import numpy as np


# 타입에 따라 데이터 분리 및 인덱스 저장
numerical_indices = [i for i, x in enumerate(players) if isinstance(x, int)]
numerical_data = [x for x in players if isinstance(x, int)]

categorical_indices = [i for i, x in enumerate(players) if isinstance(x, str)]
categorical_data = [x for x in players if isinstance(x, str)]

# 숫자 데이터 스케일링
if numerical_data:
    numerical_data = np.array(numerical_data).reshape(-1, 1)
    scaler = StandardScaler()
    scaled_numerical_data = scaler.fit_transform(numerical_data).flatten()
else:
    scaled_numerical_data = []

# 문자 데이터 인코딩
if categorical_data:
    categorical_data = np.array(categorical_data).reshape(-1, 1)
    encoder = OrdinalEncoder()
    encoded_categorical_data = encoder.fit_transform(categorical_data).flatten()
else:
    encoded_categorical_data = []

# 원래 순서대로 결과를 재구성
processed_data = [None] * len(players)
for i, value in zip(numerical_indices, scaled_numerical_data):
    processed_data[i] = value

for i, value in zip(categorical_indices, encoded_categorical_data):
    processed_data[i] = value

# 결과 출력
print("Original Data:", players)
print("Processed Data:", processed_data)

```
