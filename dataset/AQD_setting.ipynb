```python
import pandas as pd

bronze = pd.read_csv('Athena/RANKED_SOLO_5x5_BRONZE_I.csv', nrows=900)
gold = pd.read_csv('Athena/RANKED_SOLO_5x5_GOLD_I.csv', nrows=900)
grandmaster = pd.read_csv('Athena/RANKED_SOLO_5x5_GRANDMASTER_I.csv', nrows=900)
iron1 = pd.read_csv('Athena/RANKED_SOLO_5x5_IRON_I.csv', nrows=900)
iron2 = pd.read_csv('Athena/RANKED_SOLO_5x5_IRON_II.csv', nrows=900)
silver = pd.read_csv('Athena/RANKED_SOLO_5x5_SILVER_I.csv', nrows=900)

bronze.shape, gold.shape, grandmaster.shape, iron1.shape, iron2.shape, silver.shape
```

```python
bronze['tier'] = 'BRONZE_I'
gold['tier'] = 'GOLD_I'
grandmaster['tier'] = 'GRANDMASTER_I'
iron1['tier'] = 'IRON_I'
iron2['tier'] = 'IRON_II'
silver['tier'] = 'SILVER_I'
```

```python
silver.head(7)
```

```python
data = pd.concat([bronze, gold, grandmaster, iron1, iron2, silver], axis=0)
```

```python
data = data.replace({True: 1, False: 0})

data.head()
```

```python
data.drop(['matchId', 'gameMode', 'gameVersion', 'role'], axis=1, inplace=True)

data.head()
```

```python
data.to_csv('RIOT_data.csv')
```
